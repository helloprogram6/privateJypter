```python
import pandas as pd

data = pd.read_csv("1_feature_selected.csv",nrows = 500,header=None)
data
```

```python
import numpy as np
data[0] = np.where(data[0]==37,1,0)
```

```python
x = data.iloc[:,1:].values
y = data[0].values
```

```python
x.values
```

```python
import torch
import torch.nn.functional as F     # 激励函数都在这
# 101 -- 1024 是训练数据，1--100是测试数据
trainX = torch.Tensor(x)
trainY = torch.LongTensor(y)
```

```python
trainY
```

```python
trainX
```

```python
class RNN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.rnn=torch.nn.LSTM(
            input_size=23,
            hidden_size=64,
            num_layers=2,
            batch_first=True
        )
        self.out=torch.nn.Linear(in_features=64,out_features=2)

    def forward(self,x):
        # 以下关于shape的注释只针对单向
        # output: [batch_size, time_step, hidden_size]
        # h_n: [num_layers,batch_size, hidden_size] # 虽然LSTM的batch_first为True,但是h_n/c_n的第一维还是num_layers
        # c_n: 同h_n
        print(x)
        output,(h_n,c_n)=self.rnn(x)
        print(output.size())
        # output_in_last_timestep=output[:,-1,:] # 也是可以的
        output_in_last_timestep=h_n[-1,:]
        # print(output_in_last_timestep.equal(output[:,-1,:])) #ture
        x=self.out(output_in_last_timestep)
        return x
userNet = RNN() # 10维的特征，100个隐藏节点，输出4个类别
```

```python
userNet
```

```python
# 训练代码
loss_fn = torch.nn.CrossEntropyLoss() # 损失函数
optimizer = torch.optim.Adam(userNet.parameters(),lr=3e-14)#优化器使用Adam,学习率为0.05

for t in range(1000):
    out = userNet(trainX)     # 喂给 net 训练数据 x, 输出分析值

    loss = loss_fn(out, trainY)     # 计算两者的误差
    if t % 100 == 0:
        print("Epoch",t,loss.item())
    optimizer.zero_grad()   # 清空上一步的残余更新参数值
    loss.backward()         # 误差反向传播, 计算参数更新值
    optimizer.step()        # 将参数更新值施加到 net 的 parameters 上
```
