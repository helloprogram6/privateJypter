```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,50)
y = x**2+1
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2

plt.figure()#第一张图
plt.plot(x,y1)

plt.figure(num=3,figsize=(8,5))#第二张图
plt.plot(x,y2)
plt.plot(x,y1,color="red",linewidth=3,linestyle="--")#增加红色的虚线

plt.show()
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2

plt.figure(num=3,figsize=(8,5))#第二张图
plt.plot(x,y2)
plt.plot(x,y1,color="red",linewidth=3,linestyle="--")#增加红色的虚线
#限制取值范围,及设置标签
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel("I am X")
plt.ylabel("I am Y")
#自定义坐标
new_ticks = np.linspace(-1,2,5)
plt.xticks(new_ticks)
plt.yticks([-2,-1.8,-1,1.22,3],
          ["$really\ bad$","bad","normal","good","really good"])
#设置坐标
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.spines["bottom"].set_position(("data",-1))
ax.spines["left"].set_position(("data",0))

plt.show()
```

```python
plt.figure()
# 画图例
l1, = plt.plot(x,y2,label="up")
l2, = plt.plot(x,y1,color="red",linewidth=3,linestyle="--",label="best")#增加红色的虚线
plt.legend(handles=[l1,l2,],labels=["a","b"],loc="best")
#plt.legend(handles=[l1,],labels=["a",],loc="best") #只打印第一条线
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)
plt.plot(x, y,)
#移动坐标轴的位置
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# 注解
x0 = 1
y0 = 2*x0 + 1
plt.scatter(x0,y0,s=50,color="b")#蓝色的点

plt.plot([x0,x0],[y0,0],"k--",lw=2.5)#黑色的虚线
#增加箭头的注释
plt.annotate("$2x+1=%s$"%y0,xy=(x0,y0),xycoords="data",xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

#增加红字注释
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 16, 'color': 'r'})
plt.show()
```
